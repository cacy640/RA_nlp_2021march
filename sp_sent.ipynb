```python
#########=======tokenization=================##########

import csv
import pandas as pd
from textblob import TextBlob

infile = '/Users/jiaxuan/Dropbox/sp_results/sample/osp_v2_lem_rmstopwords_covid.csv'

bloblist = list()
sentencelist = list()

with open(infile, 'r', encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    next(rows)
    for row in rows:
        text = row[3]
        blob_text = TextBlob(text)
        sentence = blob_text.sentences
        sentencelist.append(sentence)

osp_tokenized = pd.DataFrame(sentencelist)

osp_tokenized.to_csv('/Users/jiaxuan/Dropbox/sp_results/sample/osp_v2tokenized_lem_rmstopwords_covid.csv', index = False)
```

```python
########=========sentiment===================#########

import csv
import pandas as pd
from textblob import TextBlob

infile = '/Users/jiaxuan/Dropbox/sp_results/sample/osp_v2tokenized_lem_rmstopwords_covid.csv'
osp_tokenized_textblob_polarity = pd.DataFrame()

# osp polarity

bloblist = list()
i = 0

for i in range(410):
    with open(infile, 'r', encoding="utf-8") as csvfile:
        rows = csv.reader(csvfile)
        next(rows)
        bloblist = list()
        for row in rows:
            sentence = row[i] #477
            blob = TextBlob(sentence)
            bloblist.append((blob.sentiment.polarity))
        osp_tokenized_textblob_polarity[str(i)] = bloblist
        print('column ' + str(i) + ' is done.')
        i += 1 

osp_tokenized_textblob_polarity.to_csv('/Users/jiaxuan/Dropbox/sp_results/sample/osp_v2tokenized_lem_rmstopwords_covid_textblob_polarity.csv', index = False)
```
